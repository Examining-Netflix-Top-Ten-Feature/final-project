```python
df_full['as_of_date'] = pd.to_datetime(df_full['As of'])
df_full['netflix_release_date'] = pd.to_datetime(df_full['Netflix Release Date'])
df_full['Netflix Exclusive'] = df_full['Netflix Exclusive'].fillna('NO')
df_full['release_year'] = df_full['release_year'].astype('Int64')
```

```python
df_full.info(verbose=True)
```

```python
#feature engineer for the first full dataframe (which uses the netflix titles csv -- ie not IMDB)
df_full['age_days_since_rel'] = df_full['as_of_date'] - df_full['netflix_release_date']
```

# Export to CSV


```python
#export data to csv for this first full dataframe (not the imdb dataframe)
df_full.to_csv('./DATA/netflix_and_titles_ds1.csv', sep=',', encoding='utf-8')
```

```python
df_full['country'].fillna(0, inplace=True)
df_full['listed_in'].fillna(0, inplace=True)
df_full.head()
```

```python
# Splitting the genres into separate columns at commas
genres_split = df_full['listed_in'].str.split(', ', expand=True)

# Creating binary columns for each unique value
genres_dummies = pd.get_dummies(genres_split.stack()).groupby(level=0).sum()

# Concatenating the binary columns with the original DataFrame
df_full = pd.concat([df_full, genres_dummies], axis=1)

# Drop Listed in
# df_full.drop('listed_in', axis=1, inplace=True)

#
# df_full
```

```python
# Extract numerical values and units into separate columns
df_full['duration_minutes'] = df_full['duration'].str.extract(r'(\d+) min', expand=False).astype(float)
df_full['duration_seasons'] = df_full['duration'].str.extract(r'(\d+) Season[s]?', expand=False).astype(float)

# Fill NaN values in case there are no seasons or minutes mentioned
df_full['duration_minutes'].fillna(0, inplace=True)
df_full['duration_seasons'].fillna(0, inplace=True)
```

```python
# Splitting the country into separate columns at commas
country_split = df_full['country'].str.split(', ', expand=True)

# Creating binary columns for each unique value
country_dummies = pd.get_dummies(country_split.stack()).groupby(level=0).sum()

# Concatenating the binary columns with the original DataFrame
df_full = pd.concat([df_full, country_dummies], axis=1)
```
